```python
import torch
from tensor_fusion.net import ARF_with_AR_TextSubNet

input_sizes = (300, 5, 20)
hidden_sizes = (128, 32, 32)
fusion_size = 128
out_size = 1
dropouts = (0, 0, 0, 0, 0)
device = 'cpu'
dtype = torch.float32

tt_model = ARF_with_AR_TextSubNet(input_sizes, hidden_sizes, fusion_size, [0.1, 100], out_size, dropouts, device=device, dtype=dtype)
```

```python
from tensor_fusion.dataset import get_cmu_mosi_dataset

# get dataset
train_set, valid_set, test_set = get_cmu_mosi_dataset(binary=True, device=device, dtype=dtype)
```

```python
from tensor_fusion.train import binary_map_train_ARF_with_AR_TextSubNet

coeff = 1e-4
lr = 1e-3

tt_max_accuracy, tt_max_model, tt_results = binary_map_train_ARF_with_AR_TextSubNet(
    tt_model, train_set, test_set, coeff, lr, print_result=True)
```

```python
import numpy as np
print(np.mean(tt_results['train_time']))
```

```python
import time
text, audio, vision, label = test_set[:2]
tic = time.time()
output = tt_max_model(text, audio, vision)
toc = time.time()
print("TT Model Inference Time: {:.4f}".format(toc - tic))
```
